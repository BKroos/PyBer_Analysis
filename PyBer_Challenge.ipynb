### 4.3 Loading and Reading CSV files

```python
# Add Matplotlib inline magic command
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import matplotlib as mpl
import pandas as pd

# File to Load (Remember to change these)
city_data_to_load = "resources/city_data.csv"
ride_data_to_load = "resources/ride_data.csv"

# Read the City and Ride Data
city_data_df = pd.read_csv(city_data_to_load)
ride_data_df = pd.read_csv(ride_data_to_load)
```

### Merge the DataFrames

```python
# Combine the data into a single dataset
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])

# Display the data table for preview
pyber_data_df.head()
```

## Deliverable 1: Get a Summary DataFrame 

```python
#  1. Get the total rides for each city type
urban_cities_df = pyber_data_df[pyber_data_df["type"] == "Urban"]
suburban_cities_df = pyber_data_df[pyber_data_df["type"] == "Suburban"]
rural_cities_df = pyber_data_df[pyber_data_df["type"] == "Rural"]
urban_ride_count = urban_cities_df.count()["ride_id"]
suburban_ride_count = suburban_cities_df.count()["ride_id"]
rural_ride_count = rural_cities_df.count()["ride_id"]
#ride_count = pyber_data_df.groupby(["type"]).count()["ride_id"]
#ride_count
```

```python
# 2. Get the total drivers for each city type
urban_drivers = city_data_df[city_data_df['type'] == 'Urban'].sum()['driver_count']
suburban_drivers = city_data_df[city_data_df['type'] == 'Suburban'].sum()['driver_count']
rural_drivers = city_data_df[city_data_df["type"] == 'Rural'].sum()['driver_count']
#suburban_drivers = city_data_df.groupby(["Suburban"]).sum()["driver_count"]
#rural_drivers = city_data_df.groupby(['Rural']).sum()["driver_count"]
#drivers = pyber_data_df.groupby(["type"]).sum()["driver_count"]
```

```python
#  3. Get the total amount of fares for each city type
urban_fares_total = urban_cities_df.sum()["fare"]
suburban_fares_total = suburban_cities_df.sum()["fare"]
rural_fares_total = rural_cities_df.sum()["fare"]
#total_fares = pyber_data_df.groupby(["type"]).sum()["fare"]
#total_fares
```

```python
#  4. Get the average fare per ride for each city type. 
urban_fares_avg = urban_cities_df.mean()["fare"]
suburban_fares_avg = suburban_cities_df.mean()["fare"]
rural_fares_avg = rural_cities_df.mean()["fare"]
#avg_fares = pyber_data_df.groupby(["type"]).mean()["fare"]
#avg_fares
```

```python
# 5. Get the average fare per driver for each city type. 
urban_fare_per_driver = urban_fares_total / urban_drivers
suburban_fare_per_driver = suburban_fares_total / suburban_drivers
rural_fare_per_driver = rural_fares_total / rural_drivers
```

```python
#  6. Create a PyBer summary DataFrame. 
pyber_summary_df = pd.DataFrame(data=
                               {'Total Rides':[urban_ride_count, suburban_ride_count, rural_ride_count],
                               'Total Drivers': [urban_drivers, suburban_drivers, rural_drivers],
                               'Total Fares': [urban_fares_total, suburban_fares_total, rural_fares_total],
                               'Average Fare per Ride': [urban_fares_avg, suburban_fares_avg, rural_fares_avg],
                               'Average Fare per Driver': [urban_fare_per_driver, suburban_fare_per_driver, rural_fare_per_driver]},
                               index=['Urban', 'Suburban', 'Rural'])

#pyber_summary_df = pd.DataFrame([{"Total Rides":[rural_ride_count, suburban_ride_count, urban_ride_count], 
                                #"Total Drivers":drivers, 
                                #"Total Fares":total_fares, 
                                #"Average Fare per Ride":avg_fares, 
                                #"Average Fare per Driver":[rural_fare_per_driver, suburban_fare_per_driver, urban_fare_per_driver]}], index=["Rural", "Suburban", "Rural"])
pyber_summary_df
```

```python
#  7. Cleaning up the DataFrame. Delete the index name
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
#  8. Format the columns.
pyber_summary_df["Total Fares"] = pyber_summary_df["Total Fares"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Ride"] = pyber_summary_df["Average Fare per Ride"].map("${:,.2f}".format)
pyber_summary_df["Average Fare per Driver"] = pyber_summary_df["Average Fare per Driver"].map("${:,.2f}".format)
pyber_summary_df["Total Drivers"] = pyber_summary_df["Total Drivers"].map("{:,}".format)
```

## Deliverable 2.  Create a multiple line plot that shows the total weekly of the fares for each type of city.

```python
# 1. Read the merged DataFrame
pyber_data_df.head()
```

```python
# 2. Using groupby() to create a new DataFrame showing the sum of the fares 
#  for each date where the indices are the city type and date.
pyber_weekly = pyber_data_df.groupby(["type", "date"]).sum()["fare"]
pyber_weekly
```

```python
# 3. Reset the index on the DataFrame you created in #1. This is needed to use the 'pivot()' function.
pyber_weekly = pyber_weekly.reset_index()
pyber_weekly
```

```python
# 4. Create a pivot table with the 'date' as the index, the columns ='type', and values='fare' 
# to get the total fares for each type of city by the date. 
pyber_weekly_pivot = pyber_weekly.pivot(index='date', columns='type', values='fare')
pyber_weekly_pivot
```

```python
# 5. Create a new DataFrame from the pivot table DataFrame using loc on the given dates, '2019-01-01':'2019-04-29'.
pyber_q1_2019 = pyber_weekly_pivot.loc['2019-01-01':'2019-04-29']
```

```python
# 6. Set the "date" index to datetime datatype. This is necessary to use the resample() method in Step 8.
pyber_q1_2019.index = pd.to_datetime(pyber_q1_2019.index)
```

```python
# 7. Check that the datatype for the index is datetime using df.info()
pyber_q1_2019.info()
```

```python
# 8. Create a new DataFrame using the "resample()" function by week 'W' and get the sum of the fares for each week.
pyber_q1_2019_weekly = pyber_q1_2019.resample('W').sum()
pyber_q1_2019_weekly
```

```python
# 8. Using the object-oriented interface method, plot the resample DataFrame using the df.plot() function. 
pyber_q1_2019_weekly.plot(kind='line', figsize=(25, 8), title="Total Fare by City Type", xlabel="Date", ylabel="Fare ($USD)")
#mpl.rcParams['font.size'] = 14

# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')
plt.savefig("analysis/PyBerFareSummary.png")
```
